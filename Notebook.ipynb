# Data Analysis and Visualizations

```python
df.head()
```

```python
cancellation_perc = df["is_canceled"].value_counts(normalize = True)
cancellation_perc
```

```python
plt.figure(figsize=(5,4))
plt.bar(["Not Cancelled","Cancelled"],df["is_canceled"].value_counts(),color=["g","r"],edgecolor="black")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x="hotel",hue="is_canceled",data=df,palette="Blues")
plt.xlabel("Hotels")
plt.ylabel("Number of Reservations")
plt.legend(["Not Cancelled", "Cancelled"])
```

Cancellation percentage of Resort Hotel

```python
resort_hotels = df[df["hotel"]=="Resort Hotel"]
resort_hotels["is_canceled"].value_counts(normalize=True)
```

Cancellation percentage of City Hotel

```python
city_hotels = df[df["hotel"]=="City Hotel"]
city_hotels["is_canceled"].value_counts(normalize=True)
```

```python
df.head()
```

```python
resort_hotels = resort_hotels.groupby("reservation_status_date")[["adr"]].mean()
city_hotels = city_hotels.groupby("reservation_status_date")[["adr"]].mean()
```

```python
plt.figure(figsize=(14,10))
plt.title("Average Daily Rate in City and Resort Hotel",fontsize=20)
plt.plot(resort_hotels.index,resort_hotels["adr"],label="Resort Hotel")
plt.plot(city_hotels.index,city_hotels["adr"],label="City Hotel")
plt.legend(fontsize=20)
plt.show()
```

```python
df["month"] = df["reservation_status_date"].dt.month
plt.figure(figsize=(14,10))
plt.title("Reservation Status per month",fontsize=20)
ax1 = sns.countplot(x="month",hue="is_canceled",data=df)
plt.xlabel("Month")
plt.ylabel("No. of Reservations")
plt.legend(["Not Canceled","Canceled"],fontsize=15)
plt.show()
```

This shows :
- Cancellation is maximum in january and minimum in august and september
- No cancellation is maximum as well in august and minimum in January and December

```python
plt.figure(figsize=(14,8))
plt.title("ADR per month",fontsize=20)
sns.barplot(x="month",y="adr",data=df[df["is_canceled"]==1].groupby("month")[["adr"]].sum().reset_index(),hue="adr")
plt.show()
```

This proves our Hypothesis that the `cancellations are maximum` whenever the `Average Daily Rate is High`

```python
cancelled_reservations = df[df["is_canceled"]==1]
top_10country=cancelled_reservations["country"].value_counts()[:10]
```

```python
plt.figure(figsize=(8,8))
plt.pie(top_10country,autopct="%0.2f%%",labels=top_10country.index)
plt.show()
```

```python
mkt_segment = df["market_segment"].value_counts().reset_index()
mkt_segment.columns = ["Market","count"]
plt.figure(figsize=(16,8))
sns.barplot(x="Market",y="count",data=mkt_segment,hue="Market")
plt.show()
```

##### Arranging data in Ascending order by Date with respect to ADR

```python
cancelled_df_adr = cancelled_reservations.groupby("reservation_status_date")[["adr"]].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values("reservation_status_date",inplace=True)

not_cancelled_reservations = df[df["is_canceled"]==0]
not_cancelled_df_adr = not_cancelled_reservations.groupby("reservation_status_date")[["adr"]].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values("reservation_status_date",inplace=True)
```

```python
plt.figure(figsize=(20,4))
plt.title("Average Daily Rate")
plt.plot(not_cancelled_df_adr["reservation_status_date"],not_cancelled_df_adr["adr"],label="Not Canceled Reservations")
plt.plot(cancelled_df_adr["reservation_status_date"],cancelled_df_adr["adr"],label="Canceled Reservations")
plt.legend()
plt.show()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr["reservation_status_date"]>"2016") & (cancelled_df_adr["reservation_status_date"]<'2017-09')]

not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr["reservation_status_date"]>"2016") & (not_cancelled_df_adr["reservation_status_date"]<'2017-09')]
```

```python
plt.figure(figsize=(20,4))
plt.title("Average Daily Rate",fontsize=20)
plt.plot(not_cancelled_df_adr["reservation_status_date"],not_cancelled_df_adr["adr"],label="Not Canceled Reservations")
plt.plot(cancelled_df_adr["reservation_status_date"],cancelled_df_adr["adr"],label="Canceled Reservations")
plt.legend()
plt.show()
```
